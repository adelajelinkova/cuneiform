```python
import matplotlib.pyplot as plt
from detecto.utils import read_image
sample_image = "/proc_05-05_NBC11595_KWtif-R002-C006_png.rf.c7ac104e799b1751799efa3b69aa99e6.jpg"
image = read_image(bp+sample_image)
plt.imshow(image)
plt.show()
```

```python
from detecto.core import Dataset
dataset = Dataset(bp)
valset = Dataset(vp)
testset = Dataset(tp)
```

```python
from detecto.visualize import show_labeled_image
image, targets = dataset[1]
show_labeled_image(image, targets['boxes'], targets['labels'])
```

```python
from detecto.core import DataLoader, Model
labels = ['hstroke']
model = Model(labels)
```

```python
print(len(dataset), len(valset), len(testset))
```

```python
from detecto import core
loader = core.DataLoader(dataset, batch_size=4, shuffle=True)
```

```python
losses = model.fit(dataset=loader, val_dataset=valset, epochs=50, lr_step_size=3)
```

```python
import matplotlib as plt
plt.plot(losses)
plt.show()
```

```python
model.save('gdrive/MyDrive/UGS/Neural_Networks/ResNet/Resnet_weights_50epoch.pth')
```
